```python

import matplotlib.pyplot as plt
from scipy import stats

measured_dist = [0.4, 0.5, 0.6, 0.7, 0.8,]

actual_dist = [0.63, 0.74, 0.855, 0.975, 1.08]

slope, intercept, r_value, p_value, std_err = stats.linregress(actual_dist, measured_dist)

# Calculate the best fit line
y_pred = [slope * xi + intercept for xi in actual_dist]

print(f'Slope: {slope}', f'Intercept: {intercept}', f'R-squared: {r_value**2}', sep='\n')

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(actual_dist, measured_dist, color='blue', label='Actual data')  # Plot the actual data points
plt.plot(actual_dist, y_pred, color='red', label='Best fit line')  # Plot the best fit line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression: Actual Data Points and Best Fit Line')
plt.legend()
plt.show()
```
